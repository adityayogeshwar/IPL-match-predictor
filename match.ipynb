### Replacing teams old names with new names

```python
match_df['team1'] = matches['team1'].str.replace('Delhi Daredevils', 'Delhi Capitals') \
                                    .str.replace('Royal Challengers Bangalore', 'Royal Challengers Bengaluru') \
                                    .str.replace('Kings XI Punjab', 'Punjab Kings') \
                                    .str.replace('Deccan Chargers', 'Sunrisers Hyderabad')

match_df['team2'] = matches['team2'].str.replace('Delhi Daredevils', 'Delhi Capitals') \
                                    .str.replace('Royal Challengers Bangalore', 'Royal Challengers Bengaluru') \
                                    .str.replace('Kings XI Punjab', 'Punjab Kings') \
                                    .str.replace('Deccan Chargers', 'Sunrisers Hyderabad')

match_df['toss_winner'] = matches['toss_winner'].str.replace('Delhi Daredevils', 'Delhi Capitals') \
                                    .str.replace('Royal Challengers Bangalore', 'Royal Challengers Bengaluru') \
                                    .str.replace('Kings XI Punjab', 'Punjab Kings') \
                                    .str.replace('Deccan Chargers', 'Sunrisers Hyderabad')
```

#### Removing matches in which DLS used

```python
match_df=match_df[match_df['method']!='D/L']
```

#### Merging both dataFrames on match id 

```python
match_df= match_df[['id','city','winner','target_runs']]
```

```python
data = match_df.merge(deliveries,left_on='id',right_on='match_id')
```

```python
data.drop('match_id',axis=1,inplace=True)
```

#### Fetching only 2nd inning data

```python
data=data[data['inning']==2]
```

```python
data.shape
```

```python
data.head(2)
```

#### Making two new columns runs left and balls left 

```python
# Calculating current score
data['current_score']=data.groupby('id')['total_runs'].cumsum()
```

```python
data.head()
```

```python
# calculating runs required 
data['runs_required'] = data['target_runs']-data['current_score']
data.head()
```

#### Calculating number of balls left after every ball

```python
max_balls = 120  # Initialize the total number of balls

# Check if the delivery is either wides or noballs
curr_id = 335982
data['balls_left'] = max_balls 
for index, row in data.iterrows():
    if curr_id==row['id']:

        if row['extras_type'] in ['wides', 'noballs']:  # Check for wide or no-ball in the row
            data.at[index, 'balls_left'] = max_balls  # Do not decrement if wide or no-ball
        else:
            max_balls -= 1 
            data.at[index, 'balls_left'] = max_balls 
    else:
        max_balls=120
        curr_id = row['id']
        if row['extras_type'] in ['wides', 'noballs']:  # Check for wide or no-ball in the row
            data.at[index, 'balls_left'] = max_balls  # Do not decrement if wide or no-ball
        else:
            max_balls -= 1 
            data.at[index, 'balls_left'] = max_balls 


```

#### Calculating wicket left 

```python
data.columns
```

```python
wickets=data.groupby('id')['is_wicket'].cumsum().values
data['wickets_left'] = 10 - wickets
```

```python
data.head(2)
```

#### Creating columns like current run rate and required run rate

```python
data['crr'] = (data['current_score']*6/(120 - data['balls_left'])).round(2)
data['rrr'] = (data['runs_required']*6/(data['balls_left'])).round(2)
```

```python
data.head(1)
```

##### Creating column win 

```python
data['result'] = (data['winner'] == data['batting_team']).astype(int)
```

```python
data.head(2)
```

```python
# Taking only those columns which are required 
df=data[['city','batting_team','bowling_team','target_runs','current_score','runs_required','balls_left','wickets_left','crr','rrr','result']]
```

```python
df.shape
```

```python
df.head(2)
```

#### Shuffling data to make randomness in the data

```python
df = df.sample(df.shape[0])
```

```python
df.head(5)
```

```python
df.shape
```

#### Checking null values

```python
print(df.shape)
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
print(df.shape)
df.isnull().sum()
```

```python
df.describe()
```

#### Removing last ball data because it is creating a error that infinite will at required run rate come as 0 balls left and some runs are required 

```python
df=df[df['balls_left']!=0]
```

##### Also removing those first balls when a wide or no ball is bowled

```python
df=df[df['balls_left']!=120]
```

```python
df.describe()
```

```python
x = df.iloc[:,0:-1]
y = df.iloc[:,-1]
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
print("The shape of X_train is ",X_train.shape)
print("The shape of X_test is ",X_test.shape)
print("The shape of Y_train is ",Y_train.shape)
print("The shape of Y_test is ",Y_test.shape)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer(transformers=[
    ('trf', OneHotEncoder(sparse_output=False, drop='first'), ['batting_team', 'bowling_team', 'city'])
], remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
pipe.fit(X_train,Y_train)
```

```python
y_pred = pipe.predict(X_test)
from sklearn.metrics import accuracy_score
accuracy_score(Y_test,y_pred)
```

#### Checking for SVM

```python
from sklearn.svm import SVC
pipe2 = Pipeline(steps=[
    ('step1',trf),
    ('step2',SVC(kernel='rbf', C=1.0, gamma='scale'))
])
pipe2.fit(X_train,Y_train)
```

```python
y_pred2 = pipe2.predict(X_test)
accuracy_score(Y_test,y_pred2)
```

#### Checking whole match condition after each over

```python
def match_progression(data,match_id,pipe):
    match = data[data['id'] == match_id]

    match = match[(match['ball'] == 6)]
    temp_df = match[['batting_team','bowling_team','city','runs_required','balls_left','wickets_left','target_runs','crr','rrr','current_score']].dropna()
    temp_df = temp_df[temp_df['balls_left'] != 0]
    result = pipe.predict_proba(temp_df)
    temp_df['lose'] = np.round(result.T[0]*100,1)
    temp_df['win'] = np.round(result.T[1]*100,1)
    temp_df['end_of_over'] = range(1,temp_df.shape[0]+1)
    
    target = temp_df['target_runs'].values[0]
    runs = list(temp_df['runs_required'].values)
    new_runs = runs[:]
    runs.insert(0,target)
    temp_df['runs_after_over'] = np.array(runs)[:-1] - np.array(new_runs)
    wickets = list(temp_df['wickets_left'].values)
    new_wickets = wickets[:]
    new_wickets.insert(0,10)
    wickets.append(0)
    w = np.array(wickets)
    nw = np.array(new_wickets)
    temp_df['wickets_in_over'] = (nw - w)[0:temp_df.shape[0]]
    
    print("Target-",target)
    temp_df = temp_df[['end_of_over','runs_after_over','wickets_in_over','lose','win']]
    return temp_df,target
```

```python
temp_df,target = match_progression(data,335982,pipe)
temp_df
```
